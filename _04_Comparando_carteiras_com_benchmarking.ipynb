```
#Importando e instalando bibliotecas
!pip install yfinance --upgrade --no-cache-dir
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as web
import yfinance as yf
yf.pdr_override()

import seaborn as sns
```

```
tickerspc=["SUZB3.SA","KLBN11.SA"] #Ativos do setor de papel e celulose
tickersba=["SANB4.SA","ITUB4.SA"] #Ativos do setor financeiro
tickersav=["AZUL4.SA","GOLL4.SA"] #Ativos do setor de aviação
tickersdiv=["SUZB3.SA","SANB4.SA","AZUL4.SA"] #Ativos diversificados


carteira1=web.get_data_yahoo(tickerspc,period="3y")["Adj Close"]
carteira2=web.get_data_yahoo(tickersba,period="3y")["Adj Close"]
carteira3=web.get_data_yahoo(tickersav,period="3y")["Adj Close"]
carteira4=web.get_data_yahoo(tickersdiv,period="3y")["Adj Close"]

ibov=web.get_data_yahoo("^BVSP",period="3y")["Adj Close"]
```

```
carteira1,carteira2,carteira3,carteira4
```

```
ibov
```

```
carteira1=carteira1.dropna()
carteira2=carteira2.dropna()
carteira3=carteira3.dropna()
carteira4=carteira4.dropna()

ibov=ibov.dropna()

carteira1,carteira2,carteira3,carteira4
```

```
sns.set()
sns.set_style("ticks")
carteira1.plot(figsize=(20,8))
carteira2.plot(figsize=(20,8))
carteira3.plot(figsize=(20,8))
carteira4.plot(figsize=(20,8))
sns.set_context("notebook", font_scale=1.1, rc={"lines.linewidth": 1.2})
plt.ylabel('R$')
plt.xlabel("Data")
```

```
carteira1_norm=(carteira1/carteira1.iloc[0])*1000
carteira2_norm=(carteira2/carteira2.iloc[0])*1000
carteira3_norm=(carteira3/carteira3.iloc[0])*1000
carteira4_norm=(carteira4/carteira4.iloc[0])*667
carteira1_norm, carteira2_norm, carteira3_norm,carteira4_norm
```

```
carteira1_norm["Saldo"]=carteira1_norm.sum(axis=1)
carteira2_norm["Saldo"]=carteira2_norm.sum(axis=1)
carteira3_norm["Saldo"]=carteira3_norm.sum(axis=1)
carteira4_norm["Saldo"]=carteira4_norm.sum(axis=1) 
carteira1_norm, carteira2_norm,carteira3_norm,carteira4_norm
```

```
ibov_norm=(ibov/ibov.iloc[0])*2000
ibov_norm
```

```
sns.set()
sns.set_style("ticks")
sns.set_context("notebook", font_scale=1.1, rc={"lines.linewidth": 1.2})
carteira1_norm["Saldo"].plot(figsize=(20,8),label="Carteira Papel e Celulose")
carteira2_norm["Saldo"].plot(figsize=(20,8),label="Carteira Bancos")
carteira3_norm["Saldo"].plot(figsize=(20,8),label="Carteira Aviação")
carteira4_norm["Saldo"].plot(figsize=(20,8),label="Carteira Diversificada")
ibov_norm.plot(label="Ibovespa")

plt.legend()
plt.xlabel('Data')
plt.ylabel('R$')


```
