* Demonstração de Valor Adicionado (DVA)

# **Acessando a base de dados**

```python
url_base='http://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/ITR/DADOS/'
```

```python
#Criando uma lista com todos os nomes dos arquivos
arquivos_zip=[]
for ano in range(2011,2021):
  arquivos_zip.append(f'itr_cia_aberta_{ano}.zip')

arquivos_zip
```

```python
#Fazendo download de arquivos
for arq in arquivos_zip:
  wget.download(url_base+arq)
```

```python
#Extraindo os arquivos ZIP
for arq in arquivos_zip:
  ZipFile(arq,'r').extractall('CVM')
  
```

```python
#Criando uma pasta com dados históricos
nomes=['BPA_con','BPA_ind','BPP_con','BPP_ind','DFC_MD_con','DFC_MD_ind','DFC_MI_con','DFC_MD_ind','DMPL_con','DMPL_ind','DRE_con','DRE_ind','DVA_con','DVA_ind']
for nome in nomes:
   arquivo=pd.DataFrame()
   for ano in range(2011,2021):
     arquivo=pd.concat([arquivo,pd.read_csv(f'CVM/itr_cia_aberta_{nome}_{ano}.csv',sep=';',decimal=',',encoding='ISO-8859-1')])
   arquivo.to_csv(f'Dados/itr_cia_aberta_{nome}_2011-2021', index=False)

```
